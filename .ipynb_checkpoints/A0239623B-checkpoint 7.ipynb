```python
import pandas as pd
import numpy as np
```

# Reading Data 

```python
train = pd.read_csv("Assignment2_train.csv")
test = pd.read_csv("Assignment2_test.csv")
```

```python
train.head()
```

```python
# drop index column 
train = train.drop(columns=["index"])
```

```python
train.head()
```

Import Label Encoder - To turn categorical variables into numericals

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
cat_cols = ['race','sex','education','f1','f2','f3','f4','f5']
train[cat_cols] = train[cat_cols].apply(label_encoder.fit_transform)
```

```python
train.head()
```

```python
# target data 
y = train['y']

# features 
x = train.iloc[:,:-1]

```

```python
# check for null values 
x.isnull().sum()
# do not have missing values 
```

```python
from sklearn.model_selection import train_test_split
# split dataset into training set and validation set
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.3, random_state=23)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
```

# Logistic Regression 

```python

from sklearn.linear_model import LogisticRegression
```

```python
# instantiate log reg model
log_model = LogisticRegression(random_state=101)
```

```python
# fit the model with data 
log_model.fit(x_train,y_train)
y_val_logmodel = log_model.predict(x_val)
print("Accuracy:", metrics.accuracy_score(y_val,y_val_logmodel))
print("Recall:", metrics.recall_score(y_val,y_val_logmodel))
print("Precision:", metrics.precision_score(y_val,y_val_logmodel))
print("F1 Score:", metrics.f1_score(y_val,y_val_logmodel))

# confusion matrix 
cfn_matrix = metrics.confusion_matrix(y_val,y_val_logmodel,labels=[1,0])
print("\nConfusion Matrix:\n")
print(cfn_matrix)

# Classification Report 
print("\nClassification Report:\n")

print(classification_report(y_val, y_val_logmodel))
```

```python
#plot roc curve 
import matplotlib.pyplot as plt 
# predicted probabilities of class 1 
y_pred_prob_logmodel = log_model.predict_proba(x_val)[:,1]
fpr, tpr, _ = metrics.roc_curve(y_val, y_pred_prob_logmodel)
auc = metrics.roc_auc_score(y_val, y_pred_prob_logmodel)
plt.plot(fpr,tpr,label="LogReg_Model, auc ="+str("{:.3f}".format(auc)))
plt.legend(loc=4)
plt.show()
```

Grid Search to get the best parameters 


```python
from sklearn.model_selection import GridSearchCV

grid_params = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': np.logspace(-4,4,20),
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'max_iter': [90,100]
}

gridCV = GridSearchCV(log_model, param_grid=grid_params, verbose=True)
gridCV.fit(x_train,y_train)
```

```python
print(gridCV.best_params_)
```

```python
good_log_model = LogisticRegression(C=545.5594781168514, max_iter = 100, penalty='l2', solver = "newton-cg", random_state=101)
good_log_model.fit(x_train,y_train)
y_val_goodlogmodel = good_log_model.predict(x_val)
print("Accuracy:", metrics.accuracy_score(y_val,y_val_goodlogmodel))
print("Recall:", metrics.recall_score(y_val,y_val_goodlogmodel))
print("Precision:", metrics.precision_score(y_val,y_val_goodlogmodel))
print("F1 Score:", metrics.f1_score(y_val,y_val_goodlogmodel))
```

```python
#plot roc curve 
import matplotlib.pyplot as plt 
# predicted probabilities of class 1 
y_pred_prob_goodlogmodel = good_log_model.predict_proba(x_val)[:,1]
fpr, tpr, _ = metrics.roc_curve(y_val, y_pred_prob_goodlogmodel)
auc = metrics.roc_auc_score(y_val, y_pred_prob_goodlogmodel)
plt.plot(fpr,tpr,label="LogReg_Model, auc ="+str("{:.3f}".format(auc)))
plt.legend(loc=4)
plt.show()
```
